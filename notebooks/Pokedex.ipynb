```python
import pandas as pd

# Paths
pokemon_csv_path = '/Users/thomasstycke/Desktop/DTU/MLOps/MLOps-project100/data/raw/pokemon.csv'
pokelist_path = '/Users/thomasstycke/Desktop/DTU/MLOps/MLOps-project100/data/raw/pokemon_list.txt'

# Read the list of Pokémon names from pokelist.txt
with open(pokelist_path, 'r') as f:
    pokemons = [line.strip() for line in f.readlines()]

# Load the pokemon.csv file
df = pd.read_csv(pokemon_csv_path, sep='\t', encoding='utf-16-le')

# Filter the DataFrame to keep only rows where 'english_name' is in the list of Pokémon names
filtered_df = df[df['english_name'].str.lower().isin([pokemon.lower() for pokemon in pokemons])]
filtered_df.head()
```

```python
# Which pokemons are in the list but not in the dataset?
missing_pokemons = set(pokemons) - set(filtered_df['english_name'].str.title())
print(missing_pokemons)
```
